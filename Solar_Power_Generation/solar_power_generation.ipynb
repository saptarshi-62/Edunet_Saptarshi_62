## Solar power generation from a given dataset

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Reading the dataset
df=pd.read_csv("solarpowergeneration.csv")
# Showing the first few rows of the dataset
df.head()
```

```python
# Showing some rows from the last
df.tail()
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Checking the details of each and every attribute's datatype of the dataset
df.info()
```

```python
# Checking for the shape of the dataset
df.shape
```

```python
# Checking if there are any duplicayed values in the dataframe
df.duplicated().sum()
```

```python
# Getting some insights about the dataset
df.describe()
```

```python
# Importing matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# Plot distribution of power 
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=40)
plt.title("Distribution of Generated Power (KW)")
plt.xlabel("Generated Power")
plt.ylabel("Frequency")
plt.show()
```

## EDA

```python
df[df.columns[:9]].hist(bins=30,figsize=(20,10))
plt.show()
```

```python
df[df.columns[9:18]].hist(bins=30,figsize=(20,10))
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=30,figsize=(8,4))
plt.show()
```

```python
# Bivariate Analysis
# Scatter plot with the target feature
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    if(column!='generated_power_kw'):
        plt.subplot(7,3,i+1)
        plt.scatter(df[column],df['generated_power_kw'])
        plt.title(f'{column} vs Generated Power (KW)')
        plt.xlabel(column)
        plt.ylabel('Generated Power (KW)')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True,fmt='.2f')
plt.show()
```

```python
# Outliers
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.show()
```
