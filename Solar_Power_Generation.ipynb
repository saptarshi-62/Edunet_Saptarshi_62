## Solar power generation from a given dataset

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Reading the dataset
df=pd.read_csv("solarpowergeneration.csv")
#Showing the first few rows of the dataset
df.head()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
#Checking the details of the dataset
df.info()
```

```python
# Checking if there are any duplicayed values in the dataframe
df.duplicated().sum()
```

```python
# Importing matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# Plot distribution of power 
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30)
plt.title("Distribution of Generated Power (KW)")
plt.xlabel("Generated Power")
plt.ylabel("Frequency")
plt.show()
```
